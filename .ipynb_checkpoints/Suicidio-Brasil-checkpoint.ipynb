### Tendncia da taxa de suicdio no Brasil


Uma primeira comparao que farei diz respeito ao nmero de suicdios cometidos no Brasil por 100 mil habitantes, e a tendncia do grfico em relao  taxa mundial.

Para ver a tendncia,  melhor sempre usar essa relao por 100 mil, pois a populao do pas cresceu muito de 1985 para 2015. Se formos comparar em termos absolutos, poderemos tirar concluses erradas ou distorcidas.

```python
# pegar a media mundial e do Brasil em suicidios
years = dados_Brasil.year.unique()    # pegar os anos para o eixo x
suicides_brasil_mean = dados_Brasil.groupby('year')['suicides/100k pop'].mean()
suicides_world_mean = dados.groupby('year')['suicides/100k pop'].mean()

# como o Brasil nao tem 2016, vou eliminar do dataframe mundial essa entrada
suicides_world_mean.drop(2016, inplace=True)

# plotar lineplot comparativo entre Brasil e Mundo
ax = sns.lineplot(x=years, y=suicides_brasil_mean, label='Brasil')
ax = sns.lineplot(x=years, y=suicides_world_mean, label='Mundo')
plt.legend(title="Taxa de suicdio")
plt.show()
```

Uma coisa que  facilmente percebida no grfico acima  que apesar da taxa de suicdios no Brasil ser menor que a mdia mundial, ela vem crescendo constantemente ao longo de 30 anos.

A linha laranja teve um pico por volta de 1995, porm reverteu o slope e vem caindo ano aps ano. J a linha azul (Brasil) tem um slope positivo praticamente ao longo de todo o perodo analisado.

### Faixa etria com maior ndice de sucidio

Abaixo, eu criei uma tabela dinmica (pivot table) para analisar as 6 faixas etrias em funo do ano e do nmero de suicdios por 100 mil habitantes.


Meu principal objetivo aqui  identificar qual a faixa etria que tem a maior representatividade entre aqueles que tiram a prpria vida, e identificar se houve alguma mudana no perfil ao longo de 30 anos.

```python
# criar uma tabela dinmica
table = pd.pivot_table(dados_Brasil, values='suicides_no', index=['year'], columns=['age'])

# reordenar as tableas para deixar em ordem crescente
column_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years']
table = table.reindex(column_order, axis=1)

# ver as 5 primeiras entradas da pivot table
table.head()
```

### Plotar a pivot table das faixas etrias

```python
table.plot.bar(stacked=True, figsize=(16,8))
plt.legend(title='Idade')
plt.show()
```

Mesmo visualmete  fcil identificar que o grupo de pessoas que mais cometem suicdio est entre 35-54 anos. Em segundo lugar, esto pessoas entre 25-34 anos.

Juntos, esses dois grupos correspondem a quase 60% dos registros do banco de dados.

```python
#Estatistica de suicdios por faixa etria.

table[['25-34 years']].loc[2015]
table.loc[2015] / table.loc[2015].sum()
```

Um outro ponto que vale a pena ressaltar  o aumento de casos entre pessoas acima de 55 anos. Em 1985, as pessoas com mais idade representavam uma pequena fatia do nmero total.

Entretanto, em 2015  ntido que mesmo para pessoas acima de 75 anos houve um incremento significativo no nmero de suicdios.

Para se inferir a causa dessa mudana de padro,  necessrio se avaliar mais profundamente questes que vo alm dos nmeros (como exemplo, fatores qualitativos, momento econmico do pas e a cultura dominante de cada poca).

### Taxa de suicdio entre homens e mulheres


Uma outra anlise de grande importncia  ver a porcentagem dos suicdios entre homens e mulheres.

Analisando-se todo o perodo, o dataset utilizado mostrou que aproximadamente 78% dos casos foram cometidos por homens e 22% deles por mulheres. Optou-se por pegar a mdia dos 30 anos, pois no houve mudana significativa desse comportamento durante o perodo.

```python
#Extrair valores entre homens e mulheres

homem_mulher = dados_Brasil.groupby('sex').suicides_no.sum()/ dados_Brasil.groupby('sex').suicides_no.sum().sum()
```

### Criar um grafico  entre homens e mulheres

```python
plt.pie(homem_mulher, labels=['mulheres','homens'], autopct='%1.1f%%', shadow=True)

plt.show()
```

### Correlaes entre o PIB, IDH e nmero de suicdios


Criando uma matriz de correlao e plotando um heatmap, infere-se que o aumento no PIB per capita no diminuiu o nmero de suicdios por 100 mil habitantes. Na verdade, ele se manteve estvel, contrariando o senso comum da maioria das pessoas.

Em relao ao IDH, como foi mencionado anteriormente, h muitos valores ausentes nas clulas, o que pode dar uma interpretao incorreta ou com vis.
